```python
import numpy as np
import matplotlib.pyplot as plt

# Create a time array from 0 to 2 seconds, with small steps
time = np.linspace(0, 2, 1000)

# Define the sine wave and exponential decay
frequency = 10  # Hz
decay_rate = 3  # how fast it fades

# Create the signal
signal = np.exp(-decay_rate * time) * np.sin(2 * np.pi * frequency * time)

# Plot it
plt.plot(time, signal)
plt.title("Exponentially Decaying Sine Wave")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```
